```python
sns.set_style('dark')
import warnings
warnings.filterwarnings("ignore")
```

# Runs Scored over the years

```python
fig,ax = plt.subplots()

df = dataframe.set_index('Start Date').resample('Y').agg({'Runs':'sum'})

sns.lineplot(
    data = df,
    y = 'Runs',
    x = df.index,
    marker = 'o'
)

ax.set_xlabel('Date')
ax.set_xticklabels(df.index.year)
ax.set_title("Runs scored between 2008 - 2017",fontweight = 'bold')

plt.show()
```

#### Conclusion
* Kholi scored maximum runs in the year 2012

# Distribution of matches played as per batting order

```python
df = (dataframe
      .groupby(['Pos'])
      .agg(count = ('Pos','count'))
      .sort_values(by = ['count'],ascending=[False])
      )

df_sum = df[2:].sum().to_frame().T
df_sum.index = ['Others']
df_sum

pie_df = pd.concat([df[:2],df_sum])
pie_df

fig,ax = plt.subplots()

ax.pie(
    x = pie_df['count'],
    startangle = 90,
    colors = sns.color_palette('husl'),
    explode = [0.1,0,0],
    autopct = '%.0f%%'
)

ax.legend(pie_df.index,bbox_to_anchor = (1.1,1))
ax.set_title('Matches played at batting order',fontweight = 'bold')

plt.show()
```

#### Conclusion
* Kholi Played most of his matches at number 3

# Runs scored at different batting orders

```python
fig,ax = plt.subplots()

sns.barplot(
    data = dataframe,
    y = 'Runs',
    x = 'Pos',
    palette='husl',
    estimator = 'sum',
    errorbar = None
)

ax.set_xlabel("Batting order")
ax.set_title("Runs scored at different batting orders",fontweight = 'bold')

plt.show()
```

#### Conclusion
* Kholi has scored more than 4000 runs at number 3

# Comparison of runs based on innings

```python
fig,ax = plt.subplots()

sns.boxplot(
    data = dataframe,
    y = 'Runs',
    x = 'Inns',
    palette='husl',
)

ax.set_title("Distribution of runs based on innings",fontweight = 'bold')
ax.set_xlabel("Innings")
ax.set_xticklabels(['1st Innings','2nd Innings'])

plt.show()
```

#### Conclusion
* Kholi has better average in 2nd innings than in 1st Innings

# Comparison of runs scored against countries

```python
df = (dataframe
 .groupby(['Opposition'])
 .agg({'Runs':'sum'})
 ).sort_values(by  =['Runs'],ascending=[False])[::-1]

fig,ax = plt.subplots()

ax.barh(
    df.index,
    df['Runs'],
    color = sns.color_palette('husl')
)

ax.set_title("Runs scored Vs Opponents",fontweight = 'bold')
ax.set_xlabel("Runs")

plt.show()
```

#### Conclusion
* Kholi has scored most runs against Sri Lanka

# Exploring relationship of 4s and 6s with runs

```python
fig,ax = plt.subplots(2,1,figsize = (8,6))

sns.regplot(
    data = dataframe,
    x = 'Runs',
    y = '4s',
    ci = None,
    ax = ax[0]
)

ax[0].set_title("Relationship between Runs and 4s",fontweight = 'bold')

sns.regplot(
    data = dataframe,
    x = 'Runs',
    y = '6s',
    ci = None,
    ax = ax[1],
    color = 'orange'
)

ax[1].set_title("Relationship between Runs and 6s",fontweight = 'bold')

fig.tight_layout()
plt.show()
```

#### Conclusion
* There is a linear relationship between 4s and runs but no strong linear relationship as in case of 6s. 
* It means that Virat Kohli likes playing fours. The more runs he scores in the innings, the more fours he plays.

# Strike rate trends over the years

```python
fig,ax = plt.subplots()
df = (dataframe
    .set_index('Start Date')
    .resample('Y')
    .agg({'SR':'mean'})
    )

sns.lineplot(
    data = df,
    x = df.index,
    y = df['SR'],
    palette = 'husl',
    marker = 'o'
)

ax.set_title('Strike Rate Trends Over The Years',fontweight = 'bold')
ax.set_ylabel('Year')
ax.set_xlabel('Strike Rate')

plt.show()
```

#### Conclusion
* There is a significant rise in strike rate after 2012
